### 1. Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', palette='deep')

# Customize plot
plt.title('Total Bill vs Tip by Day and Time')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show plot
plt.show()
```

### 2. Line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
fmri = sns.load_dataset('fmri')

# Create line plot
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='region', style='event')

# Customize plot
plt.title('FMRI Signal Over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Region and Event')
plt.tight_layout()

# Show plot
plt.show()
```

### 3. Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
penguins = sns.load_dataset('penguins')

# Create histogram
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack', bins=20, palette='muted')

# Customize plot
plt.title('Distribution of Flipper Lengths by Species')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Count')
plt.legend(title='Species')
plt.tight_layout()

# Show plot
plt.show()
```

### 7. Heat Map

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot('month', 'year', 'passengers')

# Create heatmap
heatmap_plot = sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')

# Customize plot
plt.title('Number of Passengers Over Time')
plt.xlabel('Year')
plt.ylabel('Month')

# Show plot
plt.show()
```

### 8. Combining Plots with FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create FacetGrid
g = sns.FacetGrid(tips, col='time', row='sex', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=.7)

# Customize plot
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Total Bill vs Tip by Time and Sex')

# Show plot
plt.show()
```

### 9. Customizing and Saving Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create plot
sns.set_theme(style='whitegrid')
scatter_plot = sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time', palette='deep')

# Customize plot
plt.title('Total Bill vs Tip by Time')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# Save plot
plt.savefig('total_bill_vs_tip.png', dpi=300)

# Show plot
plt.show()
```

### 15. KDE Plot

### 16. Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset('iris')

# Create PairGrid
g = sns.PairGrid(iris, hue='species', palette='viridis')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()

# Customize plot
g.fig.suptitle('PairGrid of Iris Dataset')

# Show plot
plt.show()
```

### 17. Residual Plot

### 18. Rug Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create rug plot
sns.kdeplot(data=tips, x='total_bill', fill=True)
sns.rugplot(data=tips, x='total_bill', color='r')

# Customize plot
plt.title('Rug Plot with KDE of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')

# Show plot
plt.show()
```

### 19. Complex Plot with Multiple Elements

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset('titanic')

# Create figure and axis
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Left subplot: Bar plot
sns.barplot(data=titanic, x='class', y='survived', hue='sex', ax=ax[0], palette='coolwarm')
ax[0].set_title('Survival Rate by Class and Sex')
ax[0].set_xlabel('Class')
ax[0].set_ylabel('Survival Rate')

# Right subplot: Count plot
sns.countplot(data=titanic, x='deck', hue='class', ax=ax[1], palette='viridis')
ax[1].set_title('Passenger Count by Deck and Class')
ax[1].set_xlabel('Deck')
ax[1].set_ylabel('Count')

# Show plot
plt.tight_layout()
plt.show()
```
