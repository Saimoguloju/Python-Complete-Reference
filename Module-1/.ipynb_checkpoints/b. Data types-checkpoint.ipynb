## Datatypes

 ![python logo](https://static.javatpoint.com/python/images/python-data-types.png)

- Data Type represent the type of data present inside a variable.
- In Python we are notrequired to specify the type explicitly. Based on value provided,the type
  willbeassignedautomatically.HencePythonisDynamicallyTypedLanguage.
+ Python contains the following inbuilt data types
1. int 
2. float 
3. complex
4. bool
5. string 
6. bytes 
7. bytearray 
8. range
9. list
+ int data type:
